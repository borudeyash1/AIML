# Histogram: Transaction Amount Distribution
"

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Transaction_Amount'], bins=50, color='blue', edgecolor='black')
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")
plt.title("Distribution of Transaction Amounts")
plt.show()
```

# Bar Chart: Transaction Type Count


```python
plt.figure(figsize=(8, 5))
df['Transaction_Type'].value_counts().plot(kind='bar', color='orange')
plt.xlabel("Transaction Type")
plt.ylabel("Count")
plt.title("Number of Transactions by Type")
plt.xticks(rotation=45)
plt.show()
```

# Pie Chart: Device Type Usage


```python
plt.figure(figsize=(6, 6))
df['Device_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['blue', 'green', 'red'])
plt.title("Device Type Distribution")
plt.ylabel("")
plt.show()
```

# Scatter Plot: Transaction Amount vs. Risk Score


```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Transaction_Amount'], df['Risk_Score'], alpha=0.5, color='purple')
plt.xlabel("Transaction Amount")
plt.ylabel("Risk Score")
plt.title("Transaction Amount vs. Risk Score")
plt.show()
```

# Box Plot: Fraud vs. Non-Fraud Transaction Amounts


```python
plt.figure(figsize=(8, 5))
df.boxplot(column='Transaction_Amount', by='Fraud_Label', grid=False)
plt.xlabel("Fraud Label (0: Non-Fraud, 1: Fraud)")
plt.ylabel("Transaction Amount")
plt.title("Transaction Amount Distribution by Fraud Label")
plt.suptitle("")
plt.show()
```
