# Matplotlib and SciPy in Python

## Introduction
Matplotlib is a popular Python data visualization library used to create static, animated, and interactive plots.
SciPy is a Python library used for scientific and technical computing.

```python
# Import required libraries
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from scipy import linalg, optimize, integrate, interpolate, stats, spatial, ndimage
```

## Matplotlib Basics

```python
# Basic Line Plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 50]
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")
plt.legend()
plt.show()
```

## Types of Plots in Matplotlib

# Bar Chart

```python
# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values, color=['red', 'blue', 'green', 'purple'])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart Example")
plt.show()
```

# Scatter Plot

```python
# Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='green', marker='x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")
plt.show()
```

# Histogram

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel("Data Range")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

# Pie Chart


```python
# Pie Chart
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [30, 25, 20, 25]
colors = ['blue', 'red', 'green', 'yellow']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart Example")
plt.show()
```

## Subplots in Matplotlib


# Creating Subplots


```python
fig, axes = plt.subplots(2, 2, figsize=(8, 6))
axes[0, 0].plot([1, 2, 3], [4, 5, 6])
axes[0, 0].set_title("Line Plot")
axes[0, 1].bar(["A", "B", "C"], [3, 5, 7], color='c')
axes[0, 1].set_title("Bar Chart")
axes[1, 0].scatter([1, 2, 3], [4, 5, 6], color='r')
axes[1, 0].set_title("Scatter Plot")
axes[1, 1].hist([1, 2, 1, 2, 2, 3, 3, 3, 4, 4], bins=4, color='m')
axes[1, 1].set_title("Histogram")
plt.tight_layout()
plt.show()
```

## SciPy Functions


### Linear Algebra: Solve Linear Equations

```python
A = np.array([[3, 2], [2, 3]])
B = np.array([5, 7])
solution = linalg.solve(A, B)
print("Solution for x and y:", solution)
```

### Optimization: Minimize a Function


```python
def func(x):
    return x**2 + 3*x + 5
result = optimize.minimize(func, x0=0)
print("Minimum value of function:", result.x)
```

### Integration: Compute Definite Integral


```python
def func(x):
    return x**2
result, error = integrate.quad(func, 0, 3)
print("Integral result:", result)
```

### Interpolation: Predict Values


```python
x = np.array([0, 1, 2, 3, 4])
y = np.array([1, 3, 7, 13, 21])
f = interpolate.interp1d(x, y, kind='linear')
print("Interpolated Value at x=2.5:", f(2.5))
```

# Statistics: Compute Mean, Median, and Normal Distribution


```python
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
print("Median:", np.median(data))
normal_data = stats.norm.rvs(size=5, loc=0, scale=1)
print("Random Normal Distribution:", normal_data)
```

# 3D Plotting


```python
# Create a figure and 3D axis 
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
# Create Data 
X = np.linspace(-5, 5, 50) 
Y = np.linspace(-5, 5, 50) 
X, Y = np.meshgrid(X, Y) 
Z = np.cos(np.sqrt(X**2 + Y**2)) 
# Create 3D Surface Plot 
ax.plot_surface(X, Y, Z, cmap='viridis') 
# Labels and Title 
ax.set_xlabel("X-axis") 
ax.set_ylabel("Y-axis") 
ax.set_zlabel("Z-axis") 
ax.set_title("3D Surface Plot") 
# Show Plot 
plt.show() 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 50)
Y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Surface Plot")
plt.show()
```

```python
# Create a figure and 3D axis 
fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d') 
# Data 
x = [1, 2, 3, 4, 5] 
y = [5, 4, 3, 2, 1] 
z = np.zeros(5) 
dx = dy = np.ones(5) 
dz = [3, 6, 2, 5, 4] 
# Create 3D Bar Chart 
ax.bar3d(x, y, z, dx, dy, dz, color='cyan') 
# Labels and Title 
ax.set_xlabel("X-axis") 
ax.set_ylabel("Y-axis") 
ax.set_zlabel("Z-axis") 
ax.set_title("3D Bar Chart") 
# Show Plot 
plt.show() 
```
